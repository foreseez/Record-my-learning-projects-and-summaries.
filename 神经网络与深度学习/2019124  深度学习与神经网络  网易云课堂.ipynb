# 向量化和非向量化的区别

```python
import time 
a = np.random.rand(1000000)
b = np.random.rand(1000000)
tic = time.time()
c= np.dot(a,b)
toc = time.time()
print(c)
print("Vectorized version:" + str(1000*(toc -tic)) +"  ms") #转化成毫秒

c= 0
tic = time.time()
for i in range(1000000):    #这是用for 循环来实现对应位置相乘
    c+=a[i]*b[i]
toc = time.time()

print(c)
print("for loop :"+ str(1000*(toc-tic))+" ms")
```

### 可以从上面的结果看出，运行时间不是一个量级的。向量化之后运行速度回大幅提升

# python 中的广播 broadcasting 向量加上常数 

```python
import numpy as np 
import pandas as pd
A = np.array([[56.0,0.0,4.4,68.0],
            [1.2,104.0,52.0,8.0],
            [1.8,135.0,99.0,0.9]]) #注意括号的个数
print(A)
```

```python
cal = A.sum(axis=0)
print(cal)
percentage =100*A/(cal.reshape(1,4)) # 矩阵的每一列的数除以总和 56/59
print(percentage)
```

* （m,n） +-*/      (1,n) (m,1)
*  会扩展成(m,n)的矩阵  复制行或列
